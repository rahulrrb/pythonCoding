Sort a List without using built-in sort, min, max function

```python
list_num = [2,3,6,6,5]
for i in range(len(list_num)):
    for j in range(i+1,len(list_num)):
        if list_num[i]>list_num[j]:
            list_num[i], list_num[j] = list_num[j], list_num[i]

print(list_num)
```

Given the participants' score sheet for your University Sports Day, you are required to find the runner-up score. You are given  scores. Store them in a list and find the score of the runner-up.

**Find our Runner up Without using in Built Function - If there is duplicate elements exists**

```python
n = int(input())
scores = list(map(int, input().split()))

# First pass: Find the maximum score
max_score = scores[0]
for i in range(1, n):
    if scores[i] > max_score:
        max_score = scores[i]
# Second pass: Find the runner-up (second maximum)
runner_up = -999999  # Very small initial value
for i in range(n):
    if scores[i] > runner_up and scores[i] < max_score:
        runner_up = scores[i]

print(runner_up)
```
